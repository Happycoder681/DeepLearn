**Computer vision is the process of providing the ability to understand the context of images and videos to computer by converting them into mathematical representation.**

**The top left corner of the image is at (0,0) and when we say the size of the picture is (1024 X 720) then we have 1024 rows and 720 columns OR the height of the image is 1024 and width is 720.**

**A greyscale image has only one channel & its a 2D array. Here the number of rows and columns give the dimension of the array.**

**In colourful image or RGB image, 3 channels Red, Green and Blue with the values ranging from 0 to 255 are stacked one above another to give a 3 dimensional array.**

# Applications of Computer Vision:
Image Segmentation i.e segmenting the image based on an object
Object Detection i.e trying to find image with the help of bounding boxes.
Image Classification - Here, bounding boxes are not required.
Multi Class Classification i.e when we have more than 2 objects of different types as target.
Multi Label Classification - Here, all the objects present in the single image are given as labels.
Face Detection and Face Recognition
Optical Character Recognition.

```python
BGR_img = cv2.cvtColor(RGB_img,cv2.COLOR_RGB2BGR)
cv2_imshow(BGR_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()  # do this when in jupyter notebook
```

```python
# Brightness of Image
# new_gray_img = gray_img+90
new_gray_img = cv2.add(gray_img,+100)
cv2_imshow(new_gray_img)
cv2.waitKey(10000)
# cv2.destroyAllWindows()
```

```python
new_gray_img = cv2.convertScaleAbs(gray_img,alpha=0.5) # alpha gives the contrast factor
cv2_imshow(new_gray_img)  # Contrast Incresred
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
B,G,R = cv2.split(color_img)
cv2_imshow(B)
cv2.waitKey(0)
cv2_imshow(G)
cv2.waitKey(0)
cv2_imshow(R)
cv2.waitKey(0)
```
